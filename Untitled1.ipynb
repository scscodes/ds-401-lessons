```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Thinkful-Ed/data-science-lectures/master/lahman_pitching.csv')
```

```python
df = df[df['yearID'] >= 1950]
```

```python
plt.hist(df['BAOpp'], bins=100, )
```

```python
plt.hist(df['HR'], bins=100)
```

```python
joined = pd.concat([df['W'], df['L']])
plt.hist(joined, bins=100, edgecolor='black')
```

```python
df.head()
```

```python
df_games = df['G']

means = []
for i in range(1000):
  sample = df_games.sample(100, replace=False)
  means.append(sample.mean())

means = pd.Series(means)
```

```python
plt.xlim([min(means)-5, max(means)+5])
plt.hist(means, alpha=0.5)
plt.title('distro of games')
plt.xlabel('average games')
plt.ylabel('count')
plt.show()
```
